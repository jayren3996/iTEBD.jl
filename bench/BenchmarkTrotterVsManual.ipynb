# Benchmark The New Trotter API Against Manual iTEBD

This notebook activates the dedicated `bench/` environment and runs the benchmark harness in smoke mode so the workflow stays quick to validate in a checkout.

````julia
using Pkg

function find_repo_root()
    for candidate in (pwd(), abspath(joinpath(pwd(), "..")))
        if isfile(joinpath(candidate, "Project.toml")) && isdir(joinpath(candidate, "bench"))
            return candidate
        end
    end
    error("Could not locate the iTEBD.jl repository root from $(pwd())")
end

repo_root = find_repo_root()
bench_root = joinpath(repo_root, "bench")

Pkg.activate(bench_root)
Pkg.resolve()
Pkg.instantiate()

println("repo_root = ", repo_root)
println("bench_root = ", bench_root)
````

````julia
benchmark_script = joinpath(bench_root, "bench_trotter_vs_manual.jl")
saved_args = copy(ARGS)
empty!(ARGS)
push!(ARGS, "--smoke")

cd(repo_root) do
    include(benchmark_script)
end

empty!(ARGS)
append!(ARGS, saved_args)
````

For the full benchmark outside the notebook, run:

```julia
julia --project=bench -e 'using Pkg; Pkg.resolve(); Pkg.instantiate(); include("bench/bench_trotter_vs_manual.jl")'
```
